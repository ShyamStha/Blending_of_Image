## Belnding and Pasting Image

```python
img1 = cv2.imread('husky.jpg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB) 
img2 = cv2.imread('labrador.jpg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB) 
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
img1.shape
```

```python
img2.shape
```

```python
# Blending images of the same size
img1 = cv2.resize(img1,(1200,1200))
img2 = cv2.resize(img2,(1200,1200))
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
blended = cv2.addWeighted(img1,0.5,img2,0.5,0)
```

```python
plt.imshow(blended)
```

```python
# OVERLAY SMALL IMAGE ON THE TOP OF A LARGER IMAGE(NO BLENDING)
# Numpy reassignment
```

```python
img1 = cv2.imread('husky.jpg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB) 
img2 = cv2.imread('labrador.jpg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB) 
```

```python
img2 = cv2.resize(img2,(600,600))
```

```python
plt.imshow(img2)
```

```python
plt.imshow(img1)
```

```python
large_img = img1
small_img = img2
```

```python
x_offset = 0
y_offset = 0
```

```python
x_end = x_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[1]
```

```python
large_img[y_offset:y_end,x_offset:x_end] = small_img
```

```python
plt.imshow(large_img)
```
